```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def gaussian(x, sigma, u):
    return ((1 / (sigma * math.sqrt(2 * math.pi)))  * math.pow(math.e, -0.5 * math.pow(((x - u) / sigma), 2)))/4
```

```python
def alternative(x): 
    return (math.cos((x + 1) * math.pi) / 2.0) + 0.5;
```

```python
def sigmoid(x):
    c = 0.5
    a = 10
    return 1 / (1 + math.pow(math.e, - a *(x - c)))
```

```python
u = 1
sigma = 0.3
multipier = 3
x = np.arange(0, 1, 0.01) 
yg = [gaussian(x, sigma, u) * multipier for x in x]
ya = [alternative(x) for x in x]
ys = [sigmoid(x) for x in x]
plt.figure(figsize=(10,5))
plt.plot(x, ya)
plt.plot(x, ys)
plt.plot(x, yg)
plt.legend(["alternative", "sigmpid", "gaussian"])
plt.title("mean=1, sigma=0.3 and gaussian multipier = 3")
plt.show()
```
